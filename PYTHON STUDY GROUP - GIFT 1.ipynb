## TITLE: CUSTOMER CHURN PREDICTION

### Task 1

```python
# Create the Variables

customer_name = "Sarah Johnson"
monthly_subscription_fee = 8500
months_with_company = 10
customer_care_calls = 6
missed_payments = 2
total_monthly_data_usage = 18
minimum_expected_usage = 20
premium_subscriber = False
```

### Task 2

```python
# Calculate Annual Subscription Value

annual_subscription = monthly_subscription_fee * 12
print("Annual Subscription Value: ",annual_subscription,"(N)")
```

### Task 3

```python
# Update Missed Payments

missed_payments += 1
print("Updated Missed Payments: ",missed_payments)
```

### Task 4

```python
# Check Data Usage

data_usage_check = monthly_data_usage < minimum_expected_usage
print("Is the monthly data usage below the company's expected usage? ",data_usage_check)
```

### Task 5

```python
# Determine if Customer Complaints are high

high_complaints = customer_care_calls > 5
print("Did the Customer contact Customer Care more than five times per month? ",high_complaints)
```

### Task 6

```python
# Determine if Payment History is poor

poor_payment_history = missed_payments >= 3
print("Is the Customer's payment history poor? ",poor_payment_history)
```

### Task 7

```python
# Predict Churn Risk

high_churn_risk = high_complaints and poor_payment_history
print("Is there a high churn risk? ",high_churn_risk)
```

### Task 8

```python
# Determine Customer Retention Eligibility

customer_retention_eligibility = high_churn_risk or data_usage_check
print("Does the Customer have an high retention eligibility? ",customer_retention_eligibility)
```

### Task 9

```python
# Generate a Customer Churn Report

print("==========================================")
print("     CUSTOMER CHURN PREDICTION REPORT     ")
print("==========================================")

print("Customer Name:          ",customer_name)
print("Monthly Subscription:   ",monthly_subscription_fee)
print("Estimated Annual Value: ",annual_subscription,"(N)")
print("Months with Company:    ",months_with_company)
print("Customer Care Calls:    ",customer_care_calls)
print("Missed Payment:         ",missed_payments)
print("Monthly Data Usage:     ",total_monthly_data_usage,"GB")
print("Premium Subscriber:     ",premium_subscriber)
print("High Complaints:        ",high_complaints)
print("Poor Payment History:   ",poor_payment_history)
print("High Churn Risk:        ",high_churn_risk)
print("Retention Offer Needed: ",customer_retention_eligibility)

print("==========================================")
```
